### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv).

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
```

### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_csv(url, sep = '\t')
chipo.head()
```

### Step 4. How many products cost more than $10.00?

```python
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x[1:]))
```

```python
chipo_filtered = chipo.drop_duplicates(['item_name', 'quantity', 'choice_description'])
chipo_one_prod = chipo_filtered[chipo_filtered['quantity'] == 1]
chipo_one_prod
```

```python
chipo_one_prod.query('item_price > 10.00').item_name.nunique()
```

### Step 5. What is the price of each item?
###### print a data frame with only two columns item_name and item_price

```python
# Borrar los elementos duplicados en la columna 'quantity' e 'item_name'
chipo_filtered = chipo.drop_duplicates(['quantity','item_name'])

# seleccionar solo los productos que tengan una cantidad de 1
chipo_one_prod = chipo_filtered[chipo_filtered['quantity'] == 1]

chipo_one_prod[['item_name', 'item_price']].sort_values(by = 'item_price', ascending=False)
```

### Step 6. Sort by the name of the item

```python
chipo_one_prod[['item_name', 'item_price']].sort_values(by = 'item_name')
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
chipo_one_prod.sort_values(by = 'item_price', ascending=False).head(1)

# El producto mÃ¡s caro es Steak Salad Bowl con valor de 11.89$
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
chipo[chipo['item_name'] == 'Veggie Salad Bowl'].count()

# El producto 'Veggie Salad Bowl' fue pedido 18 veces
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
chipo[(chipo['item_name'] == 'Canned Soda') & (chipo['quantity'] > 1)].shape[0]

# El producto 'Canned Soda' fue pedido mÃ¡s de una vez 20 veces
```
